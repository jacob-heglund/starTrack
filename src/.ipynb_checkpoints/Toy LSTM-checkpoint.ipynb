```python
# this is a small LSTM that trains the model to predict the next number in a 5-number sequence
# EX: given input (1,2,3,4,5), the expected output is 6
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
Data = [[[(i+j)/100] for i in range(5)] for j in range(100)]
target = [(i+5)/100 for i in range(100)]
```

```python
# Set data as np arrays
data = np.array(Data, dtype = float)
target = np.array(target, dtype = float)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2, random_state = 4)
model = Sequential()
# number input sequences, length of input sequence, length of each vector in the input 
model.add(LSTM((1), batch_input_shape = (None, 5, 1), return_sequences = True))
model.add(LSTM((1), return_sequences = False))

model.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['accuracy'])
model.summary() 
```

```python
history = model.fit(x_train, y_train, epochs = 1000, validation_data = (x_test, y_test))
```

```python
results = model.predict(x_test)
```

```python
plt.scatter(range(20), results, c = 'r')
plt.scatter(range(20), y_test, c = 'g')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.show()
```
